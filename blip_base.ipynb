```python
import os
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# --- 强制显示 Hugging Face Hub 下载进度条 ---
os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1" # 启用并行下载器，通常更快
os.environ["HF_HUB_PROGRESS"] = "1"         # 强制显示进度条
# --- 结束设置 ---

# 1. 加载预训练的处理器和模型
print("正在加载 BLIP-Large 模型和处理器，这可能需要一些时间...")
model_name = "Salesforce/blip-image-captioning-large"

# 明确设置 use_fast=True
processor = BlipProcessor.from_pretrained(model_name, use_fast=True)
model = BlipForConditionalGeneration.from_pretrained(model_name)
print("模型和处理器加载完成。")

# 检查是否有GPU可用，并将其移动到GPU上
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print(f"模型已移动到设备：{device}")

# 2. 准备图像
image_path = "./img/img2.jpg" # <-- 请将此替换为您本地图像的实际路径
raw_image = Image.open(image_path).convert("RGB")
plt.imshow(raw_image)
print(f"已加载图像：{image_path}")

# 3. 生成字幕
print("正在生成图像字幕...")
# --- 引入 Prompt 的部分 ---
prompt = """ """ 
print(f"使用的提示语：{prompt}")
inputs = processor(raw_image, text=prompt, return_tensors="pt").to(device)
outputs = model.generate(**inputs, num_beams=4, max_length=1000, min_length=30)
caption = processor.decode(outputs[0], skip_special_tokens=True)
print(f"\n生成的图像字幕：\n{caption}")
```
